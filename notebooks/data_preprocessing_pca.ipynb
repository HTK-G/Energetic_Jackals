```python
import pandas as pd

df = pd.read_csv("dataset.csv")
```

```python
df.shape
```

```python
df.head()
```

```python
if "Unnamed: 0" in df.columns:
    df = df.drop(columns=["Unnamed: 0"])

print("Shape after removing useless index column:", df.shape)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df_no_na = df.dropna().copy()

print("Before dropna:", df.shape)
print("After dropna:", df_no_na.shape)
print(df_no_na.isnull().sum())
```

```python
full_dups = df[df.duplicated(keep=False)].copy()

full_dups.head(20) #finds out that there are completely repetitive rows
```

```python
df_no_full_dups = df_no_na.drop_duplicates().copy()

print("Before removing fully duplicated rows:", df_no_na.shape)
print("After removing fully duplicated rows:", df_no_full_dups.shape)
print("Remaining fully duplicated rows:", df_no_full_dups.duplicated().sum()) # no full repetitive rows for now
```

```python
meta_cols = ["track_id", "artists", "album_name", "track_name"]
aux_cols = ["track_genre", "popularity", "explicit"]
feature_cols = [
    "danceability",
    "energy",
    "key",
    "loudness",
    "mode",
    "speechiness",
    "acousticness",
    "instrumentalness",
    "liveness",
    "valence",
    "tempo",
    "time_signature"
]
```

```python
keep_cols = meta_cols + aux_cols + feature_cols
df_no_full_dups = df_no_full_dups[keep_cols].copy()
```

```python
df_no_full_dups["track_id"].duplicated().sum()
```

```python
dup_rows = df_no_full_dups[df_no_full_dups["track_id"].duplicated(keep=False)].sort_values("track_id")
dup_rows.head(20)
```

```python
genre_summary = (
    df_no_full_dups.groupby("track_id")["track_genre"]
    .agg(lambda x: sorted(set(x)))
    .reset_index()
)

genre_summary["num_genres"] = genre_summary["track_genre"].apply(len) # adds a new column: the number of genres included for each song
print(genre_summary.head())
print(genre_summary["num_genres"].value_counts().sort_index()) # shows the number of songs that have different number of genres
```

```python
df_unique = df_no_full_dups.drop_duplicates(subset="track_id", keep="first").copy()
print("Original shape:", df_no_full_dups.shape)
print("Deduplicated shape:", df_unique.shape)
print("Remaining duplicated track_id count:", df_unique["track_id"].duplicated().sum())
```

```python
df_clean = df_unique[(df_unique["tempo"] > 0) & (df_unique["time_signature"] > 0)].copy()
df_clean.shape # noted that there are some rows with tempo = 0 or time_signature = 0, remove those rows
```

```python
genre_summary = genre_summary.rename(columns={"track_genre": "all_genres"})
df_clean = df_clean.merge(
    genre_summary[["track_id", "all_genres", "num_genres"]],
    on="track_id",
    how="left"
) # add the number of genres to the final dataset
```

```python
df_model = df_clean[meta_cols + aux_cols + feature_cols].copy() # now the dataset is ready for modeling

print(df_model.shape)
df_model.head()
```

```python
X = df_model[feature_cols].copy()
meta_info = df_model[meta_cols + aux_cols].copy()

print(X.shape)
print(meta_info.shape)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

X_scaled.shape
```

```python
import pandas as pd

X_scaled_df = pd.DataFrame(X_scaled, columns=feature_cols, index=X.index)
X_scaled_df.head()
```

```python
df_model.shape
```

```python
df_model[feature_cols].isnull().sum()
```

```python
X_scaled.shape
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

print(X_pca.shape)
print("Explained variance ratio:", pca.explained_variance_ratio_)
print("Total explained variance:", pca.explained_variance_ratio_.sum())
```

```python
loadings = pd.DataFrame(
    pca.components_.T,
    columns=["PC1", "PC2"],
    index=feature_cols
)

print(loadings)
```

```python
df_pca = pd.DataFrame(X_pca, columns=["PC1", "PC2"], index=X.index)

df_pca["track_name"] = meta_info["track_name"].values
df_pca["artists"] = meta_info["artists"].values
df_pca["track_genre"] = meta_info["track_genre"].values
df_pca["popularity"] = meta_info["popularity"].values
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
plt.scatter(df_pca["PC1"], df_pca["PC2"], alpha=0.2, s=5)
plt.xlabel("PC1")
plt.ylabel("PC2")
plt.title("PCA Projection of Spotify Tracks")
plt.show()
```

```python
df_clean.to_csv("clean_preprocessed_dataset.csv", index=False)
df_model.to_csv("spotify_model_input.csv", index=False)
X_scaled_df.to_csv("spotify_scaled_features.csv", index=False)
df_pca.to_csv("spotify_pca_2d.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df_scaled = pd.read_csv("spotify_scaled_features.csv")
corr = df_scaled.corr()

fig, ax = plt.subplots(figsize=(10, 8))
im = ax.imshow(corr, cmap="coolwarm", aspect="auto")

ax.set_xticks(np.arange(len(corr.columns)))
ax.set_yticks(np.arange(len(corr.columns)))
ax.set_xticklabels(corr.columns, rotation=45, ha="right")
ax.set_yticklabels(corr.columns)

for i in range(len(corr.columns)):
    for j in range(len(corr.columns)):
        ax.text(j, i, f"{corr.iloc[i, j]:.2f}", ha="center", va="center", fontsize=8)

plt.colorbar(im, label="Correlation")
plt.title("Correlation Heatmap of Audio Features")
plt.tight_layout()
plt.show()
```
