# Spotify Audio Features Exploration

This notebook demonstrates the end-to-end boilerplate pipeline for the Kaggle Spotify Audio Features dataset. It loads the dataset, preprocesses audio features, learns embeddings, clusters songs with a NumPy K-Means implementation, and visualizes the resulting embedding space.

```python
from src.data.loader import load_spotify_dataset
from src.data.preprocessing import preprocess_spotify_dataframe
from src.evaluation.metrics import summarize_pipeline_metrics
from src.models.embeddings import compute_song_embeddings
from src.models.kmeans import NumpyKMeans
from src.visualization.plots import build_embedding_scatter

catalog = load_spotify_dataset(sample_size=1500)
catalog.head()
```

```python
prepared = preprocess_spotify_dataframe(catalog)
embedding_result = compute_song_embeddings(
    prepared.feature_matrix,
    method="pca",
    embedding_dim=12,
)
labels = NumpyKMeans(n_clusters=8, random_state=42).fit_predict(embedding_result.embeddings)
catalog = prepared.cleaned_frame.copy()
catalog["cluster"] = labels
metrics = summarize_pipeline_metrics(prepared.feature_matrix, embedding_result.embeddings, labels)
metrics
```

```python
figure = build_embedding_scatter(
    catalog,
    embedding_result.visualization_projection,
    cluster_labels=labels,
    title="Spotify Embedding Map",
)
figure
```
